### **Sensors data results report**
As part of the data science challenge, a test of coding aptitude was conducted, three data samples where provided. The data samples had a three different variables, one of which was duplicated.
#### **Objectives**
To develop a script which plots data samples, identify peaks, differentiate between them and returns the depth value of each peak
#### **Results and conclusions**
A script capable of the aforementioned objectives was developed, three plots where obtained using following python libraries: Pandas for data framing and file management, Mathplot.lib for data plotting, SciPy for peak finding functions. The script prints out the depth to the corresponding voltages and two plots measured against time.
